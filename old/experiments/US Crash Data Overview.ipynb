## Demonstrate ability to collect all crashes in a city or county

```python
county = "Arlington"
state = "VA"

arlington_crashes = crashes[(crashes['State']==state) & (crashes['County']==county)]
arlington_crashes[['US_Accidents_June20.csv', 'Start_Time', 'Start_Lat', 'State', 'County', 
                   'Temperature(F)', 'Weather_Condition']].head()
```

```python
city = "Richmond"
state = "VA"

arlington_crashes = crashes[(crashes['State']==state) & (crashes['City']==city)]
arlington_crashes[['US_Accidents_June20.csv', 'Start_Time', 'Start_Lat', 'State', 'City', 
                   'Temperature(F)', 'Weather_Condition']].head()
```
